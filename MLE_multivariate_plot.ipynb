```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from src.ML_estimation import calculate_my_likelihood
from src.sucess_probability import p_g
```

```python
d_g = 2481
n_g = 66871
```

```python
# Plot of the likelihood function in w_g and gamma_g
x_num_points = 100
y_num_points = 100
w_g = np.linspace(0, 0.4, x_num_points)
gamma_g = np.linspace(-2, 2, y_num_points)
W, G = np.meshgrid(w_g, gamma_g)
L = np.zeros((x_num_points, y_num_points))
for i in range(x_num_points):
    for j in range(y_num_points):
        L[i, j] = calculate_my_likelihood(d_g, n_g, p_g, norm.pdf, w_g[i], gamma_g[j])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, G, L, cmap='viridis')
ax.set_xlabel('w_g')
ax.set_ylabel('gamma_g')
ax.set_zlabel('Likelihood')
plt.show()
```

```python
# Plot of the likelihood function in a and b
x_num_points = 30
y_num_points = 100
a = np.linspace(-3, 3, x_num_points)
b = np.linspace(-5, 5, y_num_points)
A, B = np.meshgrid(a, b)
L = np.zeros((y_num_points, x_num_points))

modified_p_g = lambda x, a, b: norm.cdf(a + b * x)
for i in range(x_num_points):
    for j in range(y_num_points):
        L[j, i] = calculate_my_likelihood(d_g, n_g, modified_p_g, norm.pdf, a[i], b[j])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(A, B, L, cmap='viridis')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('Likelihood')
plt.show()
```
